```python
from print_df import print_df 
from pandas import ExcelFile
from pandas import DataFrame

import pandas as pd

#-------------------------------------
# 엑셀파일 읽은 후 데이터 전처리
#-------------------------------------
crime5 = ExcelFile('서울시 5대 범죄 발생현황 통계.xls')

sheet_names = crime5.sheet_names
df = crime5.parse(sheet_names[0])

restrict_list = list(df['자치구']) 
print_df(restrict_list)

index_dict = {}
for i, v in enumerate(restrict_list):
    index_dict[i] = v

# 쓰지 않는 열 제거
df.drop('기간', axis=1, inplace=True)
df.drop('자치구', axis=1, inplace=True)
df.drop('합계.1', axis=1, inplace=True)
df.drop('살인', axis=1, inplace=True)
df.drop('살인.1', axis=1, inplace=True)
df.drop('강도', axis=1, inplace=True)
df.drop('강도.1', axis=1, inplace=True)
df.drop('강간강제추행', axis=1, inplace=True)
df.drop('강간강제추행.1', axis=1, inplace=True)
df.drop('절도', axis=1, inplace=True)
df.drop('절도.1', axis=1, inplace=True)
df.drop('폭력', axis=1, inplace=True)
df.drop('폭력.1', axis=1, inplace=True)

# 쓰지 않는 행 제거
df.drop([0], inplace=True)
df.drop([1], inplace=True)

# 인덱스 이름 변경
df.rename(index=index_dict, inplace=True)
df.rename(columns={'합계':'발생'}, inplace=True)

print_df(df)

# 엑셀 파일로 저장
filename = "5crime.xlsx"
df.to_excel(filename, encoding='utf-8', sheet_name='서울시 5대 강력범죄 발생건수',
            index=True, index_label='자치구', header=['발생'])
```
