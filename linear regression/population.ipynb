```python
from print_df import print_df 
from pandas import ExcelFile
from pandas import DataFrame

import pandas as pd

#-------------------------------------
# 엑셀파일 읽은 후 데이터 전처리
#-------------------------------------
population = ExcelFile('서울시인구.xls')

sheet_names = population.sheet_names
df = population.parse(sheet_names[0])

restrict_list = list(df['자치구']) 
print_df(restrict_list)

index_dict = {}
for i, v in enumerate(restrict_list):
    index_dict[i] = v
    
# 쓰지 않는 열 제거
df.drop('기간', axis=1, inplace=True)
df.drop('자치구', axis=1, inplace=True)
df.drop('인구.1', axis=1, inplace=True)
df.drop('인구.2', axis=1, inplace=True)
df.drop('인구.3', axis=1, inplace=True)
df.drop('인구.4', axis=1, inplace=True)
df.drop('인구.5', axis=1, inplace=True)
df.drop('인구.6', axis=1, inplace=True)
df.drop('인구.7', axis=1, inplace=True)
df.drop('인구.8', axis=1, inplace=True)
df.drop('세대당인구',axis=1, inplace=True)
df.drop('65세이상고령자',axis=1, inplace=True)

# 쓰지 않는 행 제거
df.drop([0], inplace=True)
df.drop([1], inplace=True)
df.drop([2], inplace=True)

# 인덱스 이름 변경
df.rename(index=index_dict, inplace=True)

print_df(df)

# 엑셀 파일로 저장
filename = "population_Seoul.xlsx"
df.to_excel(filename, encoding='utf-8', sheet_name='서울시 구별 인구수', 
             index=True, index_label='자치구', header=['세대','인구'])
```
