```python
from print_df import print_df 
from pandas import ExcelFile
from pandas import DataFrame
import pandas as pd

#-------------------------------------
# 엑셀파일 읽은 후 데이터 전처리
#-------------------------------------
dwelling = ExcelFile('서울시 건축연도별 주택현황.xls')

sheet_names = dwelling.sheet_names
df = dwelling.parse(sheet_names[0])

restrict_list = list(df['자치구']) 
print_df(restrict_list)

index_dict = {}
for i, v in enumerate(restrict_list):
    index_dict[i] = v

# 가장 최근의 데이터만 필요하기 때문에 건축연도는 지우고 총 계만 보도록 한다.
df = df.iloc[:, 0:4]

# '서울시' 데이터 제거
idx_seoul = df[df['자치구'] == '서울시'].index
df = df.drop(idx_seoul)

# 단독주택 데이터만 추출
D_data = df[df['주택종류별'] != '단독주택'].index
df = df.drop(D_data)

# 쓰지않는 열 제거 
df.drop(['자치구','기간'], axis=1, inplace=True)

# 인덱스 이름 변경
df.rename(index=index_dict, inplace=True)
print_df(df)

# 엑셀 파일로 저장
filename = "D_dwelling.xlsx"
df.to_excel(filename, encoding='utf-8', sheet_name='서울시 단독주택 수', 
             index=True, index_label='자치구', header=['주택종류별','계'])
```
